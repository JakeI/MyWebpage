# ETIT HM2 H16 Aufgabe 2a (5 Punkte)

Sei $f: \R^2 \to \R,\;\; (x,y) \mapsto xy$ , und $B =\{(x,\, y) \in \R^2 :\; x^2+\frac{y^2}{2} \leq 1\}$. Bestimmen Sie dasMinimum und Maximum von $f$ auf $B$.

```python
extent = [-1.5, 1.5, -1.8, 1.8]
res = 800
x, y = np.meshgrid(np.linspace(*extent[:2], res), np.linspace(*extent[2:], res))
mask = (x**2 + y**2/2) <= 1
z = np.zeros_like(mask)
z[mask] = 1
fig, (axa, axb) = plt.subplots(1, 2, figsize=(12,6))
#axa.imshow(z, extent=extent)
img = axa.imshow(x*y, extent=extent)
fig.colorbar(img, ax=axa)

s = res//15
X, Y = x[::s, ::s], y[::s, ::s]
dx = Y
dy = X
dR = np.sqrt(dx**2 + dy**2)
dX = dx/dR
dY = dy/dR
axa.quiver(X, Y, dX, dY)

# outline
t = np.linspace(0, 2*np.pi, 300)
X = lambda t: np.cos(t)
Y = lambda t: (np.sqrt(2))*np.sin(t)
axa.plot(X(t), Y(t), linewidth=2, color="black")

# on bondary
y = np.cos(t)*np.sqrt(2)*np.sin(t)
axb.plot(t, y, linewidth=2);

minima = np.where(np.r_[True, y[1:] < y[:-1]] & np.r_[y[:-1] < y[1:], True])[0][1:]
axa.scatter(X(t[minima]), Y(t[minima]), color="red", s=[100]*len(minima))
axb.scatter(t[minima], y[minima], color="red", s=[100]*len(minima))

maxima = np.where(np.r_[True, y[1:] > y[:-1]] & np.r_[y[:-1] > y[1:], True])[0][:-1]
axa.scatter(X(t[maxima]), Y(t[maxima]), color="green", s=[100]*len(maxima))
axb.scatter(t[maxima], y[maxima], color="green", s=[100]*len(maxima))

axa.set_xlabel("$x$")
axa.set_ylabel("$y$")
axb.set_xlabel("$\\varphi$")
axb.set_ylabel("$f(x(\\varphi), y(\\varphi)) = " + \
               "f(\\cos(\\varphi), \\sqrt{2} \\sin(\\varphi))$");
```

```python
def surface(text, f, g, extent, scatter_x, scatter_y):
    x, y = np.meshgrid(np.linspace(*extent[:2], 400), 
                       np.linspace(*extent[2:], 400))
    
    sx, sy = np.array(scatter_x), np.array(scatter_y)
    
    t = np.linspace(0, 2*np.pi, 300)
    X, Y = g(t)
    
    data = [
        go.Scatter3d(x=sx, y=sy, z=f(sx, sy), mode='markers',
                   marker=dict(
                           size=12,
                           line=dict(
                               color='rgba(217, 217, 217, 0.14)',
                               width=0.5
                           ),
                           opacity=0.8
                       )
                  ),
        go.Surface(x=x, y=y, z=f(x, y)),
        go.Scatter3d(x=X, y=Y, z=f(X, Y), mode='lines',
                     line=dict(
                         color='black',
                         width=5
                     ))
    ]

    layout = go.Layout(
        title=text,
        scene=dict(
            xaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                backgroundcolor='rgb(230, 230,230)'
            ),
            yaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                backgroundcolor='rgb(230, 230,230)'
            ),
            zaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                backgroundcolor='rgb(230, 230,230)'
            )
        )
    )

    fig = go.Figure(data=data, layout=layout)
    off.plot(fig, filename="%s.html" % text)
    disp.display(disp.HTML(
        "<a style='font-size:42pt;' href='%s.html' target='_blank'>%s</a>" 
        % (text, text)))
    
surface("16_2a", lambda x, y: x*y, 
            lambda t: (np.cos(t), np.sqrt(2)*np.sin(t)), 
            [-1.5, 1.5, -1.8, 1.8], 
            np.r_[X(t[maxima]), X(t[minima])], 
            np.r_[Y(t[maxima]), Y(t[minima])]);
```

# ETIT HM2 H16 Aufgabe 2c (3 Punkte)

Gegeben sei die Funktion $g:\; \R^2 \to \R$ mit $g(x,y) = \Bigg\{ \matrix{\frac{xy}{x^2+y^2}\;\;&,&\;\; (x,y) \neq (0,0) \\ 0 &,& (x,y) = (0,0)}$. Untersuchen Sie $g$ auf Stetigkeit in $(0,0)$. Untersuchen Sie die partielle Ableitung $\p{g}{x}(0,0)$ auf Existenz und berechnen Sie sie gegebenenfalls.

```python
extent = [-1.5, 1.5, -1.8, 1.8]
res = 800
x, y = np.meshgrid(np.linspace(*extent[:2], res), np.linspace(*extent[2:], res))

fig, axa = plt.subplots(1, 1, figsize=(16,8))
img = axa.imshow(x*y/(x**2+y**2), extent=extent)
fig.colorbar(img, ax=axa);

s = res//12
X, Y = x[::s, ::s], y[::s, ::s]
dx = (Y*(X**2+Y**2) - X*Y*2*X)/((X**2+Y**2)**2)
dy = (X*(X**2+Y**2) - X*Y*2*Y)/((X**2+Y**2)**2)
dR = np.sqrt(dx**2+dy**2)
dX = dx/dR
dY = dy/dR
axa.quiver(X, Y, dX, dY)

axa.set_xlabel("$x$")
axa.set_ylabel("$y$");
```

```python
text = "16_2c"

extent = [-1.5, 1.5, -1.8, 1.8]
x, y = np.meshgrid(np.linspace(*extent[:2], 400), 
                   np.linspace(*extent[2:], 400))

data = [go.Surface(x=x, y=y, z=x*y/(x**2+y**2))]

layout = go.Layout(
    title=text,
    scene=dict(
        xaxis=dict(
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230,230)'
        ),
        yaxis=dict(
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230,230)'
        ),
        zaxis=dict(
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230,230)'
        )
    )
)

fig = go.Figure(data=data, layout=layout)
off.plot(fig, filename="%s.html" % text)
disp.display(disp.HTML(
    "<a style='font-size:42pt;' href='%s.html' target='_blank'>%s</a>" 
    % (text, text)))
```
