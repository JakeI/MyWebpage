```python
def area(expression, extent=[-2,2, -2,2]):
    x, y = np.meshgrid(np.linspace(*extent[:2], 2000), np.linspace(*extent[2:], 2000))
    A = np.zeros_like(x)
    A[expression(x, y)] = 1.0
    fig, ax = plt.subplots(1,1, figsize=(16,8))
    img = ax.imshow(A, extent=extent, origin='lower')
    #fig.colorbar(img, ax=ax)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    return fig, ax

land = lambda a: np.logical_and(a[0], land(a[1:])) if len(a) > 1 else a[0]
```

# Blatt 10 Aufgabe 4

Berechnen Sie das Integral.

$$
\iint\limits_{[0,1] \times [0,1]} (xy+y^2) d(x,y)
$$


```python
area(lambda x, y: land([0 < x, x < 1, 0 < y, y < 1]), extent=[-0.5,1.5, -0.5,1.5]);
```

# Blatt 10 Aufgabe 3

Skizzieren Sie die Mengen $B \subset \mathbb{R}^2$, und berechnen Sie jeweils den Flächeninhalt $\iint\limits_B d(x,y)$.

 * **a)** $B = \big\{ (x,y) \in \mathbb{R}^2 \;\; \big| \;\; \frac{1}{4} x^2 − 1 < y < 2 − x \big\}$
 * **b)** $B = \big\{ (x,y) \in \mathbb{R}^2 \;\; \big| \;\; y > 0 \, , \; y^2 < x < 4 − y^2 \big\}$

```python
area(lambda x, y: land([0.25*x**2-1 < y, y < 2-x]), extent=[-7,3, -2,9]);
```

```python
area(lambda x, y: land([y > 0, y**2 < x, x < 4 - y**2]), extent=[-1,5, -0.5,2.5]);
```

# ETIT 2016F Aufgabe 2b (3 Punkte)

Bestimmen Sie die Bogenlänge der Kurve

$$
\Gamma := \big\{ (x, y) \in \mathbb{R}^2 \, : \; x = \frac{1}{4} y^2 − \frac{1}{2} \ln (y) + 6 \, , \; \;1 \leq y \leq e \big\}.
$$

```python
y = np.linspace(1, np.exp(1), 300)
yy = np.linspace(0.5, 3, 500)
X = lambda y: 0.25*y**2 - 0.5*np.log(y) + 6
fig, ax = plt.subplots(1, 1, figsize=(16,8))
ax.plot(X(yy), yy, '--', linewidth= 2)
ax.plot(X(y), y, linewidth=5)
ax.set_xlabel("x")
ax.set_ylabel("y");
```

# ETIT 2016F Aufgabe 2c (4 Punkte)

Wir betrachten das Integral

$$
\int\limits^2_0 \int\limits^1_{\sqrt{y+\frac{1}{4}} − \frac{1}{2}} \frac{2e^{x^2}}{x + 1} dx\, dy

Skizzieren Sie den Integrationsbereich, vertauschen Sie die Integrationsreihenfolge und berechnen Sie damit den Wert des Integrals.
