```python
# Newton fractal for cher_eq(A) form 1a

def newton(coeff, func, res, extent):
    x, y = np.meshgrid(np.linspace(*extent[:2], res[0]), np.linspace(*extent[2:], res[1]))
    z = x+ 1j*y
    p = np.poly1d(coeff)
    q = np.polyder(p)
    w_next = lambda w: w - p(w)/q(w)
    w = np.copy(z)
    for _ in range(100):
        w = w_next(w)
    fig, ax = plt.subplots(1,1,figsize=(15,8))
    img = ax.imshow(func(w), extent=extent)
    
newton([-1,7,0,-36], np.abs, res=[1000, 1000], extent=[-1.5,1.5, -1,1])
```

```python
newton([1,0,0,1], np.angle, res=[1000, 1000], extent=[-1,1, -1,1])
```
