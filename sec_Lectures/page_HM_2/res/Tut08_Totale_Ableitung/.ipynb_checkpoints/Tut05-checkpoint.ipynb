# Grenzwert

$$
f: A \to B
\\
\Big( \lim_{x \to 0} \; f(x) = F \Big)
\Leftrightarrow
\Big( \forall \epsilon > 0: \exists x \in A: 
\big| f(x) - F \big| < \epsilon \Big)
$$

# HMII Blatt 4 Aufgabe 2

Untersuchen Sie jeweils, ob $f(x_k, y_k)$ für $(x_k, y_k) \to (0,0)\,, \;\; (x_k, y_k) \neq (0,0)$ konvergiert, und bestimmen Sie gegebenenfalls den Grenzwert.
 * **a)** $$f(x, y) = \frac{x^2+y^2}{\sqrt{x^2 + y^2 + 1} − 1}$$
 * **b)** $$f(x, y) = \frac{xy}{e^{x^2}−1}$$

```python
def data_gen(func, limits):
    X = np.linspace(-1, 1, 400)
    Y = np.linspace(-1, 1, 400)

    x, y = np.meshgrid(X, Y)

    f = func(x, y)
    
    if limits is not None:
        f[np.where(f < limits[0])] = limits[0]
        f[np.where(f > limits[1])] = limits[1] 
    
    return x, y, f


def plot(func, title="", limits=None):
    x, y, f = data_gen(func, limits)
    
    fig = plt.figure(figsize=(15,6))
    
    fig.suptitle(title, fontsize=20)
    
    axa = fig.add_subplot(121)
    im = axa.imshow(f)
    
    axa.set_xlabel("$x$")
    axa.set_ylabel("$y$")
    
    fig.colorbar(im, ax=axa)
    
    axb = fig.add_subplot(122, projection='3d')
    axb.plot_wireframe(x, y, f) 
    
    axb.set_xlabel("$x$")
    axb.set_ylabel("$y$")
    axb.set_zlabel("$f(x, y)$")
    
    return fig
```

```python
def pyplot(func, filename=None, limits=None):
    x, y, f = data_gen(func, limits)

    surface = go.Surface(x=x, y=y, z=f)
    data = [surface]

    layout = go.Layout(
        #title=title,
        scene=dict(
            xaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                backgroundcolor='rgb(230, 230,230)'
            ),
            yaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                backgroundcolor='rgb(230, 230,230)'
            ),
            zaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                backgroundcolor='rgb(230, 230,230)'
            )
        )
    )

    fig = go.Figure(data=data, layout=layout)
    if filename is None:
        return off.iplot(fig)
    else:
        return off.plot(fig, filename=filename)
```

```python
plot(lambda x, y: (x**2 + y**2)/(np.sqrt(x**2 + y**2 + 1) - 1),
     title="$(x^2+y^2) / (\\sqrt{x^2 + y^2 + 1} − 1)$");
```

```python
pyplot(lambda x, y: (x**2 + y**2)/(np.sqrt(x**2 + y**2 + 1) - 1), filename="2a.html")

disp.display(disp.HTML("<a style='font-size: 40pt' href='2a.html' target='_blank'>Plot</a>"))
```

```python
plot(lambda x, y: x*y/(np.exp(x**2)-1),
     title="$\\frac{xy}{e^{x^2}−1}$");
```

```python
plot(lambda x, y: x*y/(np.exp(x**2)-1),
     title="$\\frac{xy}{e^{x^2}−1}$ begrenzt auf [-3,3]",
     limits=(-3, 3));
```

```python
pyplot(lambda x, y: x*y/(np.exp(x**2)-1), filename="2b.html")

disp.display(disp.HTML("<a style='font-size: 40pt' href='2b.html' target='_blank'>Plot</a>"))
```

```python
pyplot(lambda x, y: x*y/(np.exp(x**2)-1), limits=(-1, 1), filename="2bi.html")

disp.display(disp.HTML("<a style='font-size: 40pt' href='2bi.html' target='_blank'>Plot</a>"))
```


# HMII Blatt 4 Aufgabe 3

Die Funktionen $f: \, \mathbb{R}^2 \to \mathbb{R}$ und $g: \, \mathbb{R}^2 \to \mathbb{R}$ sind für $(x, y) \neq (0,0)$ durch
$$f(x, y) := \frac{2x^3}{x^2+y^2}+y^2 
\;, \;\;\;\;
g(x, y) := \frac{|xy|^\frac{1}{2}}{x^2+y^2}$$
gegeben,  und  es  sei $f(0,0) := g(0,0) :=  0$.  Untersuchen  Sie  die  Funktione $f$ und $g$ auf Stetigkeit

```python
plot(lambda x, y: 2*x**3/(x**2 + y**2) + y**2, title="$f(x, y)$");
```

```python
pyplot(lambda x, y: 2*x**3/(x**2 + y**2) + y**2, filename="3i.html")

disp.display(disp.HTML("<a style='font-size: 40pt' href='3i.html' target='_blank'>Plot</a>"))
```

```python
plot(lambda x, y: np.sqrt(np.abs(x*y))/(x**2 + y**2), title="$g(x, y)$ begrentzt auf [-3,3]", limits=(-3,3));
```

```python
pyplot(lambda x, y: np.sqrt(np.abs(x*y))/(x**2 + y**2), limits=(-3,3), filename="3ii.html")

disp.display(disp.HTML("<a style='font-size: 40pt' href='3ii.html' target='_blank'>Plot</a>"))
```
