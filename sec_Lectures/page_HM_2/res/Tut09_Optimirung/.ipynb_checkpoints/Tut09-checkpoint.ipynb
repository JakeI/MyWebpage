# Taylor Reihe

$$
f(x_1,\, \cdots \, ,\, x_n) = 
\sum_{j=0}^\infty \Bigg( 
\frac{1}{j!} \bigg( \sum_{k=1}^n 
\big(\underbrace{x_k-a_k}_{\Delta x_k}\big) 
\frac{\partial}{\partial x_k'} \bigg)^j
f(x_1', \cdots, x_n')
\Bigg) \Bigg|_{
\matrix{x_n' = a_n \\ \vdots \\ x_1' = a_1}}
$$

# ETIT Blatt 8 Aufgabe 1b

Berechnen Sie das Taylorpolynom dritten Grades der Funktion $f: \mathbb{R}^2 \to \mathbb{R}\;,\;\; f(x,\, y) = \cos(x) \sin(y) e^{x−y}$, um den Entwicklungspunkt $x^0 = (0,\, 0)$.

```python
def plot2(f, df, extent=[-1,1, -1,1], **kwargs):
    x, y = np.meshgrid(np.linspace(*extent[:2], 500), 
                       np.linspace(*extent[2:], 500))

    fig = plt.figure(figsize=(15,6))
    ax = fig.add_subplot(111)

    ax.set_title("$f(x,y) = \\cos(x) \\sin(x) e^{x-y}$" + \
                 (" normalized gradient" if "normalize" in kwargs else ""))
    
    im = ax.imshow(f(x, y), extent=extent)
    fig.colorbar(im, ax=ax)

    s = (slice(None, None, 30), slice(None, None, 30))
    X, Y = x[s], y[s]
    U, V = df(X, Y)
    if "normalize" in kwargs:
        L = np.sqrt(U**2 + V**2)
        U = U/L
        V = V/L
    Q = ax.quiver(X, Y, U, V)

    #qk = plt.quiverkey(Q, 0.9, 0.9, 2, 'normalized gradient', 
    #                   labelpos='E', coordinates='figure')

    ax.set_xlabel("$x$")
    ax.set_ylabel("$y$")
    
    return fig, ax
```

```python
f = lambda x, y: np.cos(x)*np.sin(y)*np.exp(x-y)
df = lambda x, y: ((np.cos(x)-np.sin(x))*np.sin(y)*np.exp(x-y),
                   np.cos(x)*(np.cos(y)-np.sin(y))*np.exp(x-y))
plot2(f, df, normalize=True);
```

```python
def plot1b(text, lim):

    T = lambda x, y: y + x*y - y**2 - x*y**2 + y**3/3

    x, y = np.meshgrid(np.linspace(-lim, lim, 400),
                       np.linspace(-lim, lim, 400))
    data = [go.Surface(x=x, y=y, z=f(x,y), colorscale='Jet'), 
            go.Surface(x=x, y=y, z=T(x,y), colorscale='Viridis')]

    layout = go.Layout(
        #title=title,
        scene=dict(
            xaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                backgroundcolor='rgb(230, 230,230)'
            ),
            yaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                backgroundcolor='rgb(230, 230,230)'
            ),
            zaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                backgroundcolor='rgb(230, 230,230)'
            )
        ),
        updatemenus=list([
            dict(
                buttons=list([   
                    dict(
                        label="Both", method="update",
                        args = [dict(visible=[True, True]),
                                dict(title="Both")]
                    ),
                    dict(
                        label="f(x,y)", method="update",
                        args = [dict(visible=[True, False]),
                                dict(title="f(x, y)")]
                    ),
                    dict(
                        label="T(x,y)", method="update",
                        args = [dict(visible=[False, True]),
                                dict(title="T(x, y)")]
                    ),         
                ]),
                direction = 'left',
                pad = {'r': 10, 't': 10},
                showactive = True,
                type = 'buttons',
                x = 0.1,
                xanchor = 'left',
                y = 1.1,
                yanchor = 'top' 
            ),
        ])
    )

    fig = dict(data=data, layout=layout)

    off.plot(fig, filename="1b_%s.html" % text)

    disp.display(disp.HTML("""<p>
        <a style='font-size: 40pt' href='1b_%s.html' target='_blank'>
        %s</a></p>""" % (text, text)))
    
plot1b("Klein", 1)
plot1b("Groß", 5)
```

# ETIT Blatt 8 Aufgabe 2a

Bestimmen  Sie  jeweils  alle  Stellen  lokaler  Extrema  der  Funktion $f: \mathbb{R}^2 \to \mathbb{R}$ undentscheiden Sie, ob es sich dabei um Maximal- oder Minimalstellen handelt.

 * **i)** $f(x,\, y) = 2x^3 − 3xy + 2y^3 − 3$
 * **ii)** $f(x,\, y) = x^3 − 3x^2 + xy^2 − y^2$

```python
def surface(text, f, extent, scatter_x, scatter_y):
    x, y = np.meshgrid(np.linspace(*extent[:2], 400), 
                       np.linspace(*extent[2:], 400))
    
    sx, sy = np.array(scatter_x), np.array(scatter_y)
    data = [
        go.Scatter3d(x=sx, y=sy, z=f(sx, sy), mode='markers',
                   marker=dict(
                           size=12,
                           line=dict(
                               color='rgba(217, 217, 217, 0.14)',
                               width=0.5
                           ),
                           opacity=0.8
                       )
                  ),
        go.Surface(x=x, y=y, z=f(x, y))
    ]

    layout = go.Layout(
        title=text,
        scene=dict(
            xaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                backgroundcolor='rgb(230, 230,230)'
            ),
            yaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                backgroundcolor='rgb(230, 230,230)'
            ),
            zaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                backgroundcolor='rgb(230, 230,230)'
            )
        )
    )

    fig = go.Figure(data=data, layout=layout)
    off.plot(fig, filename="%s.html" % text)
    disp.display(disp.HTML(
        "<a href='%s.html' target='_blank'>%s</a>" 
        % (text, text)))
```

```python
f = lambda x, y: 2*x**3 - 3*x*y + 2*y**3 - 3
df = lambda x, y: (6*x**2 - 3*y, -3*x + 6*y**2)
fig, ax = plot2(f, df)
ax.scatter([0,0.5], [0,0.5], color="red");
```

```python
fig, ax = plot2(f, df, normalize=True)
ax.scatter([0,0.5], [0,0.5], color="red");

#surface("plot_i", f, [-1,1,-1,1], [0,0.5], [0,0.5]);
```

```python
fig, ax = plot2(f, df, extent=[0.4, 0.6, 0.4, 0.6], normalize=True)
ax.scatter([.5], [.5], color="red");
```

```python
f = lambda x, y: x**3 - 3*x**2 + x*y**2 - y**2
df = lambda x, y: (3*x**2 - 6*x + y**2, 2*x*y - 2*y)
fig, ax = plot2(f, df, extent=[-1.5,2.8,-2.3,2.3], normalize=True)
ax.scatter([0, 2, 1, 1], [0, 0, np.sqrt(3), -np.sqrt(3)], color="red");

surface("plot_ii", f, [-1.5,2.8,-2.3,2.3], 
        [0, 2, 1, 1], [0, 0, np.sqrt(3), -np.sqrt(3)]);
```

# ETIT Blatt 8 Aufgabe 3

Bestimmen Sie mit Hilfe der Multiplikatorenregel von Lagrange diejenigen Punkte $(x,\, y) \in \mathbb{R}^2$ auf der Kreislinie $x^2 + y^2 − 2x + 2y + 1 = 0$, die vom Punkt $(−1,\, 1)$ den kleinsten bzw. dengrößten Abstand haben. Geben Sie die Abstände an.

```python
plt.figure(figsize=(8, 8))
t = np.linspace(0, 2*np.pi, 200)
plt.plot(np.cos(t)+1, np.sin(t)-1, linewidth=3)
plt.scatter([-1,1], [1,-1], color="red", s=[200]*2)
x = np.linspace(-1.5, 2, 2)
plt.plot(x, -x, linewidth=3)
dx, dy = np.cos(-np.pi/4), np.sin(-np.pi/4)
a, b = np.sqrt(2)-1, np.sqrt(2)+1
plt.scatter([a*dx, b*dx], [a*dy, b*dy], color="green", s=[200]*2)
plt.grid();
```
