# HM2 Blatt 5 Aufgabe 1

Skizzieren Sie folgende Kurven und berechnen Sie deren Längen
 * **a)** $γ: [−1,1] \to \mathbb{R}^2\;,\;\; t \mapsto (t,\,|t|)$
 * **b)** $γ: [0,2\pi] \to \mathbb{R}^2\;,\;\; t \mapsto \big(t−\sin(t), \; 1 − \cos(t)\big)$
 * **c)** $z: [0,2\pi] \to \mathbb{C}\;,\;\; \varphi \mapsto z(φ) := \varphi e^{i \varphi}$
 
*Hinweis zur Berechnung der Längen:* **b)** Schreiben Sie $\cos(t) = \cos(\frac{1}{2} t + \frac{1}{2} t)$ und verwendenSie das Additionstheorem für Cosinus. **c)** Es gilt $\int \sqrt{1 +\varphi^2}\;d\varphi = \frac{1}{2}(\text{Arsinh}{\varphi} + \varphi \sqrt{1 + \varphi^2})$.

```python
t = np.linspace(-1, 1, 500)
x = t
y = np.abs(t)

plt.figure(figsize=(16,7))
plt.plot(x, y, linewidth=5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("$t \\mapsto (t, |t|)$");
```

```python
t = np.linspace(-4*np.pi, 4*np.pi, 500)
x = t - np.sin(t)
y = 1 - np.cos(t)

plt.figure(figsize=(16,7))
plt.plot(x, y, linewidth=5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("$t \\mapsto (t - \\sin(t)\;,\;\; 1 - \\cos(t) )$");
```

```python
phi = np.linspace(0, 2*np.pi, 500)
z = phi*np.exp(1j*phi)
x = np.real(z)
y = np.imag(z)

plt.figure(figsize=(16,7))
plt.plot(x, y, linewidth=5)
plt.xlabel("$Re(z)$")
plt.ylabel("$Im(z)$")
plt.title("$\\varphi \\mapsto \\varphi e^{i\\varphi}$");
```

# HM2 Blatt5 Aufgabe 2

Die Kurve $\gamma:\; [−\sqrt{3}/2,\sqrt{2}/2]\to \mathbb{R}^3$ ist gegeben durch
$$
\gamma(t) = 
\begin{pmatrix}
\arcsin(t) \\
t \\
\sqrt{1−t^2}
\end{pmatrix} \;,\;\;
t \in [−\frac{\sqrt{3}}{2},\frac{\sqrt{2}}{2}].
$$
Ist $\gamma$ eine reguläre Kurve? Berechnen Sie die Länge der Kurve $\gamma$ und bestimmen Sie die Darstellung von $\gamma$ bezüglich der Bogenlänge.

```python
t = np.linspace(-np.sqrt(3)/2, np.sqrt(2)/2)
x = np.arcsin(t)
y = t
z = np.sqrt(1-t**2)

trace = dict(x=x, y=y, z=z, 
             line=dict(color="blue", width=15),
             mode="lines", type="scatter3d")
data = go.Data([trace])
layout = dict(title="Aufgabe 2")
fig = go.Figure(data=data, layout=layout)

off.plot(fig, filename="2.html")

disp.display(disp.HTML("<a href='2.html' target='_blank' style='font-size:64pt;text-align:center;'>Plot</a>"))
```
