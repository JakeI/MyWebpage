# Das Problem

**Beweise:**
<span style="font-size: 28pt">
$$ \forall n \in \mathbb{N}: \; q(n) $$
</span>

 - $\mathbb{N} = \{1, 2, 3, \dots \}$ ist die Menge aller natürlichen Zahlen
 - $q(n)$ Aussage
 
***
 
 - Naiver Ansatz: Zeige $q(1)$, dann $q(2)$, dann $q(3)$, ... (Vollständige Fallunterscheidung = Proof by exhaustion)
 - Leider nicht möglich (es gibt unendlich viele Natürliche Zahlen)

### Vollständige Induktion
 1. _IA_: Beweise $q(1)$ 
 2. _IV, IS_: Beweise $q(n) \Rightarrow q(n+1)$

***

#### Hinweise

 - Schritt (1) ist oft offensichtlich
 - Schritt (2) bedeutet, $q(n+1)$ auf zu schreiben, $q(n)$ einzusetzen und zu einer wahren Aussage vereinfachen.


 - _IA_ = Induktionsanfang (=base case) ist die Bedingung $q(1)$
 - _IV_ = Induktionsvoraussetzung (=Induction Hypothesis) ist die Annahme $q(n)$
 - _IS_ = Induktionsschluss (~=Inductive step) ist die Beziehung $q(n) \Rightarrow q(n+1)$



#### Vorstellung
Weil $q(1)$ und $q(n) \Rightarrow q(n+1)$ gilt, gilt auch: $q(1) \Rightarrow q(2)$ und damit $q(2)$. Im nächsten schritt kann man das selbe machen usw.


$$ q(1) \Rightarrow q(2) \Rightarrow q(3) \Rightarrow q(4) \Rightarrow \dots \Rightarrow q(n) \Rightarrow q(n+1) \Rightarrow \dots $$

***

#### Sonstiges
 - Das lässt sich theoretisch auf alle abzählbar unendlichen Mengen anwenden (z.B. für $\mathbb{Z}$, $\mathbb{Q}$, $\{ \text{alle Beume} \}$ etc.) In dieser Veranstaltung bleiben es aber meist bei den Natürlichen Zahlen $\mathbb{N}$.

## HMI H 2016 1a (3 Punkte)

Zeigen Sie, dass fur jedes $ \forall n ∈ \mathbb{N}$ gilt:
$$
\sum_{k=1}^n (-1)^k k^2 = (-1)^n \frac{n(n+1)}{2}
$$


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def aufgabe1(n=50):
    k = np.arange(1, n)
    lhs = ((-1)**k*k**2).cumsum()
    rhs = (-1)**k*k*(k+1)/2

    plt.figure(figsize=(11,8))
    plt.plot(k, lhs, linewidth=5, label="$\\sum_{k=1}^n (-1)^k k^2$")
    plt.plot(k, rhs, linewidth=2, label="$(-1)^n \\frac{n(n+1)}{2}$")
    plt.xlabel("$n$")
    plt.legend(loc="upper left");
```

```python
aufgabe1()
```

## HMI H 2015 4a (5 Punkte)

Beweisen Sie, dass fur alle $\forall n ∈ \mathbb{N}$
$$
4^n \geq n^2 + 3^n
$$
gilt.

```python
def aufgabe2():
    k = np.arange(1, 50, dtype=np.int32) # verwende "nur" 32bit ganszahlen (sind hier nicht geignet)
    lhs = 4**k
    rhs = k**2 + 3**k

    fig, (axa, axb) =  plt.subplots(1, 2, figsize=(12,6))

    s = slice(0, 10)
    #axa.semilogy(k[s], lhs[s], linewidth=3, label="$4^n$")
    #axa.semilogy(k[s], rhs[s], linewidth=3, label="$n^2 + 3^n$")
    axa.plot(k[s], lhs[s], linewidth=3, label="$4^n$")
    axa.plot(k[s], rhs[s], linewidth=3, label="$n^2 + 3^n$")
    axa.legend(loc="upper left")
    axa.set_xlabel("$n$")

    axb.plot(k, lhs, linewidth=3, label="$4^n$")
    axb.plot(k, rhs, linewidth=3, label="$n^2 + 3^n$")
    axb.legend(loc="upper left");
    axb.set_title("Numerischer Müll (Diagram sinloss)")
    axb.set_xlabel("$n$")
```

```python
aufgabe2()
```

## HMI F 2015 4a (4 Punkte)

Beweisen Sie, dass fur alle $\forall n ∈ \mathbb{N}$
$$
\sum_{k=1}^n k (k+1) (k+2) = \frac{n (n+1) (n+2) (n+3)}{4}
$$
gilt.

```python
def aufgabe3(n=50):
    k = np.arange(1, n)
    lhs = (k*(k+1)*(k+2)).cumsum()
    rhs = k*(k+1)*(k+2)*(k+3)/4

    plt.figure(figsize=(11,8))
    plt.plot(k, lhs, linewidth=5, label="$\\sum_{k=1}^n k (k+1) (k+2)$")
    plt.plot(k, rhs, linewidth=3, label="$(n (n+1) (n+2) (n+3))/4$")
    plt.xlabel("$n$")
    plt.legend(loc="upper left");
```

```python
aufgabe3()
```
