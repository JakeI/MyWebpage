# Blatt 4 Aufgabe 3

Skizzieren Sie die folgenden Mengen in der komplexen Zahlenebene:

 - *(a)* $\; \; A = \{ \; z \in \mathbb{C}: \; | z + 1 + i | = | z - 3 - 3i | \;\}$
 - *(b)* $\; \; B = \{ \; z \in \mathbb{C}:  \; | z - i | \geq 1 \; \wedge \; | z - 1 - 2i | < 3 \; \}$
 - *(c)* $\; \; C = \{ \; z \in \mathbb{C}: \; \mathrm{Re} (z^2) \leq 1 \; \}$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ex3a():
    box = [-0.5, 2.5, -0.5, 2.5]
    res = 500
    x, y = np.linspace(*box[:2], res), np.linspace(*box[2:], res)
    z = (lambda x, y: x+1j*y)(*np.meshgrid(x, y))
    cost = np.abs(np.abs(z+1+1j) - np.abs(z-3-3j))
    plt.figure(figsize=(12, 6))
    plt.imshow(cost, cmap="gray", extent=box, origin="lower", interpolation="bicubic")
    plt.colorbar()
    plt.plot(x, 2 - x, color="red", label="$\mathrm{Im}(z) = 2 - \mathrm{Re}(z)$")
    plt.legend()
```

```python
ex3a()
```

```python
def ex3b():
    box = [-3, 5, -2, 6]
    res = 500
    x, y = np.linspace(*box[:2], res), np.linspace(*box[2:], res)
    z = (lambda x, y: x+1j*y)(*np.meshgrid(x, y))
    cost = np.array([0.0*np.real(z)]*3).reshape(res, res, 3)
    cost[np.abs(z-1j) >= 1, 0] = 1.0
    cost[np.abs(z-1-2j) <3, 1] = 1.0
    plt.figure(figsize=(12, 6))
    plt.imshow(cost, extent=box, origin="lower", interpolation="bicubic")
    plt.title("$|z-i| \geq 1$ in rot, $|z-1-2j| < 3$ in grün, gelb = rot+grün")
```

```python
ex3b()
```

```python
def ex3c():
    box = [-3, 3, -3, 3]
    res = 500
    x, y = np.linspace(*box[:2], res), np.linspace(*box[2:], res)
    z = (lambda x, y: x+1j*y)(*np.meshgrid(x, y))
    cost = 0.0*np.real(z) + 1.0
    cost[np.real(z**2) <= 1] = 0.0
    plt.figure(figsize=(12, 6))
    plt.imshow(cost, extent=box, cmap="gray", origin="lower", interpolation="bicubic")
```

```python
ex3c()
```

# HMI F 2016 2a (3 Punkte)

Bestimmen Sie alle $z \in \mathbb{C}$ mit


$$ \Big| \frac{z-3}{z-2} \Big| \geq 1 $$

```python
def HMIF2016_2a():
    box = [1, 4, -1.5, 1.5]
    res = 500
    x, y = np.linspace(*box[:2], res), np.linspace(*box[2:], res)
    z = (lambda x, y: x+1j*y)(*np.meshgrid(x, y))
    cost = 0.0*np.real(z) + 1.0
    cost[np.abs((z-3)/(z-2)) >= 1] = 0.0
    plt.figure(figsize=(12, 6))
    plt.imshow(cost, extent=box, cmap="gray", origin="lower", interpolation="bicubic")
```

```python
HMIF2016_2a()
```
