# Eigenschaften des Riemannschen Integrals:

<span style="font-size: 11pt">
\begin{align}
& \int_a^b f(x) \, dx = F(x)\big|_a^b &=&\; \big[ F(x) \big]_a^b = F(b) - F(a) & \text{mit} \; F(x) := \int f(x) \, dx 
\\
& \int_a^b f(x) \, dx + \int_b^c f(x) \,dx &=&\; \int_a^c f(x) \,dx &
\\
& \int_a^b f(x) \,dx &=&\; -\int_b^a f(x) \,dx &
\\
\text{Linearit√§t:} \; \; \; & \int c \big( f(x) + g(x) \big) \,dx &=&\; c \bigg(\int f(x) \,dx + \int g(x) \,dx \bigg) &
\\
\text{Substitution:}\; \; \; & \int_{u(a)}^{u(b)} f(u) \,du &=&\; \int_a^b f\big(u(x)\big) u'(x) \,dx & \text{bzw.} \; \int f \,du = \int f(u) u' \,dx 
\\
\text{Partiell:} \; \; \; & \int_a^b u(x) v'(x) \,dx &=&\; u(x) v(x) \big|_a^b - \int_a^b v(x) u'(x) \,dx & \text{bzw.} \; \int u \,dv = uv - \int v \,du 
\end{align}
</span>

```python
def plot(title, func, a, b):
    x = np.linspace(a, b, 400)
    plt.figure(figsize=(16, 7))
    opts = dict(linewidth=5)
    symbol = lambda i: "f(x)" if i == 0 else "F(x)"
    if isinstance(func, list):
        for i, f in enumerate(func):
            plt.plot(x, f(x), label="$%s$" % symbol(i), **opts)
        plt.legend()
    else:
        plt.plot(x, func(x), **opts)
    if isinstance(title, list):
        s = ""
        ENDSTR = "\\; \\; \\; ; \\; \\; \\;"
        for i, t in enumerate(title):
            s += "%s = %s %s" % (symbol(i), t, ENDSTR)
        title = s[:len(s)-len(ENDSTR)]
    disp.display(disp.HTML(
        "<span style=\"display: block; font-size: 15px; text-align: center;\">$%s$</span>" % title))
```

```python
plot(
    [
        "\\frac{x}{\\sqrt{1-x}}", 
         "\\int \\frac{x}{\\sqrt{1-x}} = -\\frac{2}{3} \\sqrt{1-x} (2+x)"
    ],
    [
        lambda x: x/np.sqrt(1-x),
        lambda x: -(2/3)*np.sqrt(1-x)*(2+x)
    ], 
    -1, 0.9)
```

```python
with np.warnings.catch_warnings():
    np.warnings.filterwarnings('ignore')
    plot(
        [
            "\\frac{1}{\\sin^2(x) \\cos^2(x)}",
            "\int \\frac{dx}{\\sin^2(x) \\cos^2(x)} = \\frac{\\tan^3(x)}{3} + 2 \\tan(x) - \\frac{1}{\\tan(x)}"
        ],
        [
            lambda x: (np.sin(x))**-2*(np.cos(x))**-4,
            lambda x: (np.tan(x))**3/3 + 2*np.tan(x) - 1/np.tan(x)
        ], 
        -np.pi/2, np.pi)
    plt.ylim([-10, 20])
```

```python
plot("x \\, \\sin(x)", lambda x: x*np.sin(x), -1.25*np.pi, 1.25*np.pi)
```

```python
plot("\\frac{x^3}{1+x^2}", lambda x: x**3/(1+x**2), -0.25, 1.25)
```

```python
plot("x^5 \, \\sqrt{1-x^6}", lambda x: x**5*np.sqrt(1-x**6), -0.25, 1)
```
