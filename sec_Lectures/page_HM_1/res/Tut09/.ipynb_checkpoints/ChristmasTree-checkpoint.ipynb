```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as disp

import ipywidgets as widgets
from ipywidgets import interact
```

```python
t = np.linspace(0, 1, 100)
x, y = np.cos(2*np.pi*t), np.sin(2*np.pi*t)
plt.plot(x, y)
```

```python
def tree(x, y, t):
    xt = np.zeros_like(x)
    yt = np.zeros_like(y)

    def line(x, y, t, a, b, u, v):
        mask = np.logical_and(a<=t, t<=b)
        s = (t[mask]-a)/(b-a)
        xt[mask] = s*v[0] + (1-s)*u[0]
        yt[mask] = s*v[1] + (1-s)*u[1]
        return xt, yt

    # t, (x(t), y(t))
    lines = [(0.0, (0, 1)), (0.3, (-1, -0.5)), (0.4, (-0.5, -0.5)),
             (0.5, (-0.5, -1)), (0.6, (0.5, -1)), (0.7, (0.5, -0.5)), 
             (0.8, (1, -0.5)), (1, (0, 1))]
    for i in range(1, len(lines)):
        last = lines[i-1]
        current = lines[i]
        xt, yt = line(xt, yt, t, last[0], current[0], last[1], current[1])
    
    return xt, yt

xt, yt = tree(x, y, t)
plt.plot(xt, yt)
```

```python
def interp(x, y, xt, yt, s):
    return s*xt + (s-1)*x, s*yt + (s-1)*y

for i in range(5):
    plt.plot(*interp(x, y, xt, yt, i/4))
```

```python
def interact_callback(s):
    plt.plot(*interp(x, y, xt, yt, s))
    plt.show()
interact(interact_callback, s=(0.0, 1.0, 0.05))
```

```python
X, Y = np.meshgrid(np.linspace(-1.5, 1.5, 200), np.linsapce(-1.5, 1.5, 200))
Xt, Yt = tree(X, Y)
```
