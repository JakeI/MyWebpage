# Blatt 9 Aufgabe 2

* **b)** Es sei $t \in (0,2 \pi)$. Ermitteln Sie die Polarkoordinaten von $$z(t) := 1âˆ’e^{it}$$.
* **c)** Gegeben sei die komplexe Zahl $$z = \cos\bigg(\frac{5\pi}{4}\bigg)+i \sin\bigg(\frac{5\pi}{4}\bigg)$$ Berechnen Sie $z^3$ und $z^{150}$.

```python
def school_coords(ax, aspect=True):
    ax.spines['left'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position('zero')
    ax.spines['top'].set_color('none')
    ax.spines['left'].set_smart_bounds(True)
    ax.spines['bottom'].set_smart_bounds(True)
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    if aspect:
        ax.set_aspect('equal', 'datalim')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 7))

school_coords(ax)

t = np.linspace(0, 2*np.pi, 200)
z = 1 - np.exp(1j*t)
ax.plot(np.real(z), np.imag(z), linewidth=5,
        label="$1-e^{it}$")
r = 2*np.sin(t/2)
phi = 0.5*(t-np.pi)
ax.plot(r*np.cos(phi), r*np.sin(phi), linewidth=3,
        label="$ \\sin(\\frac{t}{2}) e^{\\frac{i}{2}(t-\\pi)}$")
ax.legend();
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 7))
school_coords(ax)
def view_complex(z, c, t, dx, dy):
    xy = (np.real(z), np.imag(z))
    ax.annotate("", xy=xy, 
                xycoords='data', xytext=(0, 0), 
                textcoords='data',
                arrowprops=dict(arrowstyle="->",
                                connectionstyle="arc3",
                                color=c)
                )
    ax.text(s=t, x=xy[0]+dx, y=xy[1]+dy, color=c)

view_complex(np.exp(-1j*np.pi/4), "red", "$z^3$", 0, 0)
view_complex(np.exp(1j*5*np.pi/4), "blue", "$z$", 0, -0.03)
view_complex(-1j, "green", "$z^{150}$", 0.02, 0)

ax.set_xlim([-1.3, 1.3])
ax.set_ylim([-1.9, 0.7]);
```

```python
TEXT = """
        <h2>
        $
        \\sinh(x) := \\frac{e^x - e^{-x}}{2} 
        \\; \\; ; \\; \\;
        \\cosh(x) := \\frac{e^x + e^{-x}}{2} 
        $
        </h2>
       """
disp.display(disp.HTML(TEXT))

fig, ax = plt.subplots(1, 1, figsize=(10, 7))

school_coords(ax, aspect=False)

x = np.linspace(-2, 2, 400)
s = np.sinh(x)
c = np.cosh(x)

ax.plot(x, s, label="$\\sinh(x)$", linewidth=5)
ax.plot(x, c, label="$\\cosh(x)$", linewidth=3)

ax.legend();
```
