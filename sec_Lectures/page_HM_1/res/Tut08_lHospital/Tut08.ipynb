```python
fig, (axa, axb) = plt.subplots(1, 2, figsize=(15, 7))
n = np.arange(1, 100)
sn = np.cumsum((-1)**n*(1-np.cos(1/np.sqrt(n))))
axa.plot(n, sn, label="$\\sum_{n=1}^{N} (-1)^n ( 1 - \\cos(\\frac {1}{\\sqrt{n}}))$")
axa.legend()
ssn = np.cumsum((1-np.cos(1/np.sqrt(n))))
axb.plot(n, ssn, label="$\\sum_{n=1}^{N} ( 1 - \\cos(\\frac {1}{\\sqrt{n}}))$")
axb.legend()

disp.display(disp.HTML(
    "<h2>$\\sum_{n=1}^{N} \\big( -1 \\big)^n \\bigg( 1 - \\cos\\Big( \\frac 1{\\sqrt{n}}\\Big) \\bigg)$</h2>"))
```

# ETIT 2015 F 2b (5 Punkte)

Gegeben sei die reelle Potenzreihe

$$
\sum_{n=1}^\infty \bigg( \frac{2+(-1)^n}{3} + \frac 1n \bigg)^n \big( x-2 \big)^n
$$

**i)** Berechnen Sie den Konvergenzradius $R$ der Potenzreihe.

**ii)** Ermitteln Sie die Menge aller Punkte $x \in \mathbb{R}$,in denen die Reihe konvergiert.

```python
def plotab(a, b, ax):
    X = np.linspace(a, b, 400)
    def plot(NN, **kwargs):
        N = np.arange(1, NN)
        x, n = np.meshgrid(X, N)
        series = np.sum((((2+(-1)**n)/3+1/n)*(x-2))**n, axis=0)
        ax.plot(X, series, **kwargs)
    plot(10, label="series (N=10)", linewidth=11)
    plot(100, label="series (N=100)", linewidth=8)
    plot(1000, label="series (N=1000)", linewidth=5)
    plot(10000, label="series (N=10000)", linewidth=3)
    ax.legend()
disp.display(disp.HTML("<h2>$\\sum_{n=1}^{N} \\bigg( \\frac{2+(-1)^n}{3} + \\frac 1n \\bigg)^n (x-2)^n$</h2>"))
fig, (axa, axb) = plt.subplots(1, 2, figsize=(15, 7))
plotab(1.2, 2.8, axa)
plotab(0.999, 3.001, axb)
```

# Partielle Integration
Mit $u := u(x)$, $v := v(x)$, $u' = \frac{du}{dx}$ und $v' = \frac{dv}{dx}$

$$
\int u \, dv = uv - \int v \, du 
\;\;\; \Leftrightarrow \;\;\;
\int_a^b u v' \, dx = uv\bigg|_a^b - \int_a^b v u' \, dx 
$$
